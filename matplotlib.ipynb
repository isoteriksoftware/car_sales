```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist(bins=50);
```

```python
heart_disease["age"].mean()
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10, 30));
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

ax.set(title="Disease and Cholesterol", xlabel="Age", ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50["chol"].mean(), linestyle="dashed");
```

```python
over_50.head()
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex="all")

scatter = ax1.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")
ax1.set(title="Disease and Cholesterol", ylabel="Cholesterol")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["chol"].mean(), linestyle="dashed")
ax1.set_xlim([50, 80])

scatter = ax2.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter")
ax2.set(title="Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")
ax2.legend(*scatter.legend_elements(), title="Target")
ax2.axhline(over_50["thalach"].mean(), linestyle="dashed")
ax2.set_ylim([70, 200])

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
